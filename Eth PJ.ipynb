## Volume column has info of daily transaction volume.


```python
## Converting STR data to int to use as graph
def convert_volume(vol):
    if isinstance(vol, str):  # Only process string values
        multiplier = 1
        if 'K' in vol:
            multiplier = 1e3
            vol = vol.replace('K', '')
        elif 'M' in vol:
            multiplier = 1e6
            vol = vol.replace('M', '')
        elif 'B' in vol:
            multiplier = 1e9
            vol = vol.replace('B', '')
        return float(vol) * multiplier  # Return the numerical value
    return vol  # Return as is if not a string

# Apply the function to the 'Vol.' column
eth['Vol.'] = eth['Vol.'].apply(convert_volume)

```

## Relation Between Daily Transactions and Low-High Point Differences

```python



## GRAPH 1
plt.plot(eth.index, eth['High-Low Diff'], label='High-Low Difference', color='green', linewidth=1.5, alpha=0.7)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price Difference (USD)', fontsize=12)
plt.legend(loc='upper left')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(3))  # 한 달 간격
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # 연도-월 포맷



## GRAPH 2
plt.twinx()                                             ## Dividing the Y-column since the measure is different for measure 

plt.plot(eth.index, eth['Vol.'], label='Volume', color='blue', linewidth=1.5, alpha=0.5)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Volume (in Billion USD)', fontsize=12)
plt.legend(loc='upper right')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(3))  # 세달 (Quarter 간격)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # 연도-월 format 설정 


## MAIN
plt.title('Eth  Price diff & Daily transaction Relation', fontsize=16)
plt.show()
```

```python
'''
This analysis concludes that the correlation between price volatility and trading volume is not particularly strong. While trading volume is generally expected to have a close relationship with price fluctuations in the cryptocurrency market, certain instances reveal that an increase in trading volume does not necessarily translate into a rise or fall in prices.

In particular, cases of unusually high volatility, such as in May 2021, highlight the influence of external factors like regulatory policies or market sentiment, which impact trading volume independently of price movements. These findings demonstrate that the correlation between price fluctuations and trading volume in the cryptocurrency market is not always clear or consistent.

This underscores the limitations of predicting prices or determining market direction solely based on changes in trading volume. Investors and analysts must consider a broader range of factors, including external influences, to make informed decisions.
'''
```
